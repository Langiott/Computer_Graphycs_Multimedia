```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bezier_curve(control_points, n_points=100):
    t = np.linspace(0, 1, n_points)
    n = len(control_points) - 1
    curve = sum(
        (np.math.comb(n, i) * (1-t)**(n-i) * t**i)[:, None] * control_points[i]
        for i in range(n + 1)
    )
    return curve

# Esempio
control_points = np.array([[0, 0], [1, 2], [2, 2], [3, 0]])
curve = bezier_curve(control_points)

plt.plot(curve[:, 0], curve[:, 1], label="BÃ©zier Curve")
plt.scatter(*control_points.T, color='red', label='Control Points')
plt.legend()
plt.show()
```
