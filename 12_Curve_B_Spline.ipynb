```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.interpolate import make_interp_spline

def b_spline_curve(control_points, n_points=100):
    n = len(control_points) - 1
    t = np.linspace(0, n, len(control_points))
    spline = make_interp_spline(t, control_points, k=3)
    t_fine = np.linspace(0, n, n_points)
    curve = spline(t_fine)
    return curve

# Esempio
control_points = np.array([[0, 0], [1, 2], [2, 2], [3, 0]])
curve = b_spline_curve(control_points)

plt.plot(curve[:, 0], curve[:, 1], label="B-Spline Curve")
plt.scatter(*control_points.T, color='red', label='Control Points')
plt.legend()
plt.show()
```
