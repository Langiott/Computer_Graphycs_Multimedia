```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Definizione dei vertici della mesh (coordinate 3D)
vertices = np.array([
    [0, 0, 0],  # Vertice 0
    [1, 0, 0],  # Vertice 1
    [1, 1, 0],  # Vertice 2
    [0, 1, 0],  # Vertice 3
    [0.5, 0.5, 1]  # Vertice 4 (vertice superiore)
])

# Definizione delle facce (ogni faccia è una lista di indici di vertici)
faces = [
    [0, 1, 4],  # Triangolo inferiore destro
    [1, 2, 4],  # Triangolo destro
    [2, 3, 4],  # Triangolo superiore sinistro
    [3, 0, 4],  # Triangolo sinistro
    [0, 1, 2, 3]  # Base inferiore (quadrilatero)
]

# Creazione della figura e dell'oggetto 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Conversione delle facce in una lista di coordinate
poly3d_faces = [[vertices[vertex] for vertex in face] for face in faces]

# Aggiunta delle facce alla mesh
mesh = Poly3DCollection(poly3d_faces, alpha=0.5, edgecolor='k')
ax.add_collection3d(mesh)

# Impostazioni degli assi
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

# Mostra la mesh
plt.title("Esempio di Mesh 3D Corretto")
plt.show()
```
