```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def generate_sphere(density):
    """
    Genera una mesh di una sfera con una densit di facce specifica.
    """
    phi = np.linspace(0, np.pi, density)  # Angoli polari
    theta = np.linspace(0, 2 * np.pi, density)  # Angoli azimutali
    phi, theta = np.meshgrid(phi, theta)
    
    # Coordinate della sfera
    x = np.sin(phi) * np.cos(theta)
    y = np.sin(phi) * np.sin(theta)
    z = np.cos(phi)
    
    return x, y, z

# Generiamo una sfera con densit bassa e alta
fig = plt.figure(figsize=(10, 5))

# Densit bassa
ax1 = fig.add_subplot(121, projection='3d')
x, y, z = generate_sphere(density=10)
ax1.plot_wireframe(x, y, z, color='blue')
ax1.set_title("Bassa Densit di Facce")

# Densit alta
ax2 = fig.add_subplot(122, projection='3d')
x, y, z = generate_sphere(density=50)
ax2.plot_wireframe(x, y, z, color='red')
ax2.set_title("Alta Densit di Facce")

plt.show()
```
