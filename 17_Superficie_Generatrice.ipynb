```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funzione che definisce la forma della curva (ad esempio una parabola)
def f(u):
    return np.sqrt(u)

# Parametri
u = np.linspace(0, 5, 100)  # intervallo lungo l'asse z
v = np.linspace(0, 2*np.pi, 100)  # intervallo angolare
u, v = np.meshgrid(u, v)  # creazione della griglia

# Calcolare le coordinate (x, y, z) della superficie di rivoluzione
x = f(u) * np.cos(v)
y = f(u) * np.sin(v)
z = u

# Creazione del grafico 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Tracciare la superficie
ax.plot_surface(x, y, z, cmap='viridis')

# Impostazioni del grafico
ax.set_title("Superficie di Rivoluzione")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funzione che definisce la forma della curva (ad esempio una parabola)
def f(u):
    return np.sqrt(u)

# Parametri
u = np.linspace(0, 5, 100)  # intervallo lungo l'asse z
v = np.linspace(0, 2 * np.pi, 100)  # intervallo angolare
u, v = np.meshgrid(u, v)  # creazione della griglia

# Calcolare le coordinate (x, y, z) della superficie di rivoluzione
x = f(u) * np.cos(v)
y = f(u) * np.sin(v)
z = u

# Creazione del grafico 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Tracciare la superficie di rivoluzione
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6)

# Tracciare le curve generatrici (curve lungo l'asse z)
for i in range(0, len(u), 10):  # Ridurre il numero di curve per chiarezza
    ax.plot(f(u[i]) * np.cos(v), f(u[i]) * np.sin(v), u[i], color='r', linewidth=1)

# Impostazioni del grafico
ax.set_title("Superficie creata con Curve Generatrici(rosso)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
