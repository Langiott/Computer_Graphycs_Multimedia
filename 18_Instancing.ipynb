```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funzione per generare una sfera
def generate_sphere(radius=1, resolution=10):
    u = np.linspace(0, 2 * np.pi, resolution)
    v = np.linspace(0, np.pi, resolution)
    u, v = np.meshgrid(u, v)
    x = radius * np.sin(v) * np.cos(u)
    y = radius * np.sin(v) * np.sin(u)
    z = radius * np.cos(v)
    return x, y, z

# Creare una scena
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Generare istanze della sfera in posizioni diverse
for i in range(5):
    x, y, z = generate_sphere(radius=0.5, resolution=20)
    ax.plot_surface(x + 2*i, y + 2*i, z, color='b', alpha=0.6)  # Spostare ogni istanza

ax.set_title("Primitive Instancing: Sphere Replication")
plt.show()
```
