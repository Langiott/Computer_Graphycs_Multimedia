```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funzione per la sezione traslata (un cerchio)
def circle(radius=1, num_points=100):
    theta = np.linspace(0, 2*np.pi, num_points)
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    return x, y

# Parametri per il sweep
z = np.linspace(0, 5, 100)  # percorso lungo l'asse z
radius = 1

# Creare la superficie di sweep (estrusione di un cerchio lungo un percorso)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x, y = circle(radius)

# Creare il tubo per ogni valore di z
for zi in z:
    ax.plot(x, y, np.full_like(x, zi), color='r', alpha=0.6)

ax.set_title("Sweep: Tube")
plt.show()
```
