```python
import matplotlib.pyplot as plt
import math
```

```python
def dda_line(x0, y0, x1, y1, color='black'):
    # Calcolo delle differenze
    dx = x1 - x0
    dy = y1 - y0
    
    # Calcolo della pendenza
    m = dy / dx if dx != 0 else float('inf')  # Gestisce il caso di linea verticale
    
    # Variabile per la coordinata y iniziale
    y = y0
    
    # Per memorizzare i pixel disegnati
    pixels = []
    
    # Iterazione sui pixel lungo l'asse x
    for x in range(x0, x1 + 1):
        # Approssimazione dell'intero pi√π vicino
        pixel_y = int(math.floor(y + 0.5))
        pixels.append((x, pixel_y))
        y += m
    
    # Visualizzazione della linea
    for x, y in pixels:
        plt.plot(x, y, marker='o', color=color)
    
    return pixels
```

```python


# Esempio di utilizzo
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.axis([0, 20, 0, 20])

# Disegna una linea con l'algoritmo DDA
pixels = dda_line(2, 3, 15, 10, color='blue')

plt.show()

# Stampa i pixel disegnati
print("Pixel disegnati:", pixels)
```
