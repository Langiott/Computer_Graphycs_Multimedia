```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import KDTree

# Generare dei punti casuali
np.random.seed(0)
points = np.random.rand(30, 3)  # 30 punti 3D

# Creare un KDTree
tree = KDTree(points)

# Esempio: trovare i 3 punti più vicini al punto (0.5, 0.5, 0.5)
query_point = np.array([0.5, 0.5, 0.5])
distances, indices = tree.query(query_point, k=3)

# Visualizzare i punti e il punto di ricerca
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='b', label="Punti")
ax.scatter(query_point[0], query_point[1], query_point[2], color='r', label="Punto di ricerca")
ax.scatter(points[indices, 0], points[indices, 1], points[indices, 2], color='g', label="Punti più vicini")

ax.set_title("Spatial Partitioning: KD-Tree Search")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import KDTree

# Generare dei punti casuali
np.random.seed(0)
points = np.random.rand(30, 3)  # 30 punti 3D

# Creare un KDTree
tree = KDTree(points)

# Esempio: trovare i 3 punti più vicini al punto (0.5, 0.5, 0.5)
query_point = np.array([0.5, 0.5, 0.5])
distances, indices = tree.query(query_point, k=3)

# Visualizzare i punti e il punto di ricerca
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Visualizzare i punti casuali
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='b', label="Punti")

# Visualizzare il punto di ricerca in nero
ax.scatter(query_point[0], query_point[1], query_point[2], color='k', label="Punto di ricerca", s=100)

# Visualizzare i punti più vicini in verde
ax.scatter(points[indices, 0], points[indices, 1], points[indices, 2], color='g', label="Punti più vicini", s=50)

# Impostare il titolo e le etichette degli assi
ax.set_title("Spatial Partitioning: KD-Tree Search")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Aggiungere la legenda
plt.legend()

# Mostrare il grafico
plt.show()
```
