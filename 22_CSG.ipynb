```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Funzione per un cubo
def cube(center, size):
    x = np.linspace(center[0] - size, center[0] + size, 10)
    y = np.linspace(center[1] - size, center[1] + size, 10)
    z = np.linspace(center[2] - size, center[2] + size, 10)
    X, Y, Z = np.meshgrid(x, y, z)
    return X, Y, Z

# Funzione per una sfera
def sphere(center, radius, resolution=30):
    u = np.linspace(0, 2 * np.pi, resolution)
    v = np.linspace(0, np.pi, resolution)
    u, v = np.meshgrid(u, v)
    x = center[0] + radius * np.sin(v) * np.cos(u)
    y = center[1] + radius * np.sin(v) * np.sin(u)
    z = center[2] + radius * np.cos(v)
    return x, y, z

# Creare cubo e sfera
cube_x, cube_y, cube_z = cube(center=(0, 0, 0), size=1)
sphere_x, sphere_y, sphere_z = sphere(center=(0, 0, 0), radius=1)

# Creare la figura 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Visualizzare cubo e sfera
ax.scatter(cube_x, cube_y, cube_z, color='r', alpha=0.2)
ax.scatter(sphere_x, sphere_y, sphere_z, color='b', alpha=0.5)

ax.set_title("CSG: Cube and Sphere Intersection")
plt.show()
```
