```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensioni dell'immagine
width, height = 400, 400

# Frame buffer iniziale (riempito con colore di sfondo)
background_color = np.array([0.1, 0.1, 0.1])  # Grigio scuro
framebuffer = np.full((height, width, 3), background_color)

# Funzione per visualizzare il frame buffer
def display_framebuffer(buffer, title="Frame Buffer"):
    plt.imshow(buffer)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Funzione ray tracing semplificata
def ray_tracing(framebuffer):
    # Simulazione di un ray tracing (esempio semplice)
    for x in range(framebuffer.shape[1]):
        for y in range(framebuffer.shape[0]):
            # Simula un gradiente di colore
            framebuffer[y, x] = [x / framebuffer.shape[1], y / framebuffer.shape[0], 0.5]

# Visualizza il frame buffer iniziale
display_framebuffer(framebuffer, "Frame Buffer Iniziale")

# Esegui il ray tracing
ray_tracing(framebuffer)

# Visualizza il frame buffer dopo il ray tracing
display_framebuffer(framebuffer, "Frame Buffer Dopo Ray Tracing")
```
