```python
import numpy as np
import matplotlib.pyplot as plt

def hermite_curve(P0, P1, R0, R1, n_points=100):
    t = np.linspace(0, 1, n_points)
    h00 = 2 * t**3 - 3 * t**2 + 1
    h10 = t**3 - 2 * t**2 + t
    h01 = -2 * t**3 + 3 * t**2
    h11 = t**3 - t**2
    curve = (h00[:, None] * P0 + h10[:, None] * R0 + 
             h01[:, None] * P1 + h11[:, None] * R1)
    return curve

# Esempio
P0, P1 = np.array([0, 0]), np.array([1, 1])
R0, R1 = np.array([1, 2]), np.array([-1, -2])
curve = hermite_curve(P0, P1, R0, R1)

plt.plot(curve[:, 0], curve[:, 1], label="Hermite Curve")
plt.scatter(*P0, color='red', label='P0')
plt.scatter(*P1, color='blue', label='P1')
plt.legend()
plt.show()
```
